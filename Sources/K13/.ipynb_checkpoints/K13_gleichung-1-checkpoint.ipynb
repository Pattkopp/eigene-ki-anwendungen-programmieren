```python
import pygad
```

```python
# Hyperparameter

# Anzahl der Gene
num_genes = 3
# Anzahl der Generationen
num_generations = 500
# Anzahl der Chromosomen
sol_per_pop = 100
# Anzahl der Eltern
num_parents_mating = 10
# Wahrscheinlichkeit zur Mutation
mutation_percent_genes = 34

# untere Grenze der Zufallszahl
init_range_low = -10
# obere Grenze der Zufallszahl
init_range_high = 10
# Datentyp
gene_type = float
```

```python
# Fitnessfunktion

# die Gleichung lautet: -3x -4y +3z = -5

# die Parameter sind "vorgegeben"
# solution ist besonders wichtig, enthält die Liste
def fitness_func(ga_instance, solution, solution_idx):
    x = solution[0]
    y = solution[1]
    z = solution[2]
    result = -3*x - 4*y + 3*z
    fitness = abs(-5-result) * (-1)
    
    return fitness
```

```python
# lege eine Instanz an
ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       fitness_func=fitness_func,
                       sol_per_pop=sol_per_pop,
                       init_range_low=init_range_low,
                       init_range_high=init_range_high,
                       num_genes=num_genes,
                       mutation_percent_genes=mutation_percent_genes,
                       gene_type=gene_type)

# starte Evolutionszyklus
ga_instance.run()
```

```python
# ermittle Ergebnisliste, Fitnesswert und Index der Ergebnisliste (hier nicht benötigt)
solution, solution_fitness, solution_idx = ga_instance.best_solution()

print("Beste Parameter:", solution)
print("Höchster Fitnesswert:", solution_fitness)
```

```python
# Rechnung mit dem besten Ergebnis
x = solution[0]
y = solution[1]
z = solution[2]
result = -3*x - 4*y + 3*z

print(result)
```

```python
# plotte das Ergebnis
ga_instance.plot_fitness()
```
